```python
def bathrooms_float(x):
    """
    Recibe un texto con el numero de baños y devuelve el numero en formato float.

    Parameters:
    x: Corresponde al texto con el numero de baños (el numero siendo el primer caracter del texto)

    Returns:
    Devuelve el float correspondiente
    """
    try:
        y = float(x.split(' ')[0])
    except:
        y = "Nan"
    return y
```

```python
df_listings_detailed['bathrooms'] = df_listings_detailed['bathrooms_text'].apply(lambda x: bathrooms_float(x))
df_listings_detailed = df_listings_detailed[df_listings_detailed['bathrooms'].notna()]
```

```python
df_listings_detailed['host_is_superhost'] = df_listings_detailed['host_is_superhost'].apply(lambda x: util.strtobool(x))
```

```python
df_listings_detailed = df_listings_detailed.drop(columns = ['bathrooms_text'])
```

```python
def precio_a_float(x):
    """
    Recibe un precio en formato object del tipo $60 y devuelve un float.

    Parameters:
    v: Corresponde al objeto que se quiere transformar a float

    Returns:
    Devuelve el float correspondiente
    """

    if(',' in x):
        x = x.replace(',', '')

    y = float(x.split('$')[1])
    return y
```

```python
df_listings_detailed['price'] = df_listings_detailed['price'].apply(lambda x: precio_a_float(x))
```

Ahora analizaremos el tratamiento de los outliers:

```python
import matplotlib.pyplot as plt
plt.hist(df_listings_detailed['price'])
plt.show()
```

```python
xm = df_listings_detailed['price'][df_listings_detailed['price'] < 1000]
plt.hist(xm)
plt.show()
```

Vemos que hay algunos outliers muy claros. Vamos a comprobar la cantidad de outliers, para determinar un threshold adecuado, basándonos en la medida estadistica del IQR (interquartile range)

```python
from scipy.stats import iqr
iqr = iqr(df_listings_detailed['price'])
iqr
```

```python
q3 = np.percentile(df_listings_detailed['price'], 75)
q3
```

```python
outlier_threshold = q3 + (1.5 * iqr)
outlier_threshold
```

```python
df_listings_detailed['price'][df_listings_detailed['price']>outlier_threshold].count()/17252 
```

Basándonos en la medida estadística habitualmente usada de 1,5IQRs más alla del Q3, un precio más allá de 197,5 representaría un outlier. Sin embargo, en este caso eliminaríamos el 10% del dataset. Por ello, incrementaremos ligeramente el threshold a 300€.

```python
outlier_threshold = 300
df_listings_detailed['price'][df_listings_detailed['price']>outlier_threshold].count()/17252 
```

```python
df_listings_detailed = df_listings_detailed[df_listings_detailed['price'] <= outlier_threshold] 
```
