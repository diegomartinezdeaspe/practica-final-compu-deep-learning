```python
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../AirbnbData/CleanData/df_modelo_limpio.csv')
df = df.drop(columns = (['Unnamed: 0', 'id', 'neighbourhood_cleansed', 'property_type']))
df
```

```python
df['bathrooms'] = df['bathrooms'].apply(lambda x: float(x))
```

```python
df = df.dropna()
```

```python
df.dtypes
```

### Datos

```python
X = df.drop(columns = (['price']))
y = df['price']
```

```python
labelencoder = LabelEncoder()

X['host_is_superhost'] = labelencoder.fit_transform(X['host_is_superhost'])
X['host_identity_verified'] = labelencoder.fit_transform(X['host_identity_verified'])
X['neighbourhood_group_cleansed'] = labelencoder.fit_transform(X['neighbourhood_group_cleansed'])
X['room_type'] = labelencoder.fit_transform(X['room_type'])
```

```python
CONTINUAL_COLUMNS = ['bedrooms', 'beds', 'number_of_reviews', 'bathrooms', 'puntuacion_media']
```

### Modelo

```python
input_num = tf.keras.layers.Input(shape = (5,), name = "num_input")
input_cat1 = tf.keras.layers.Input(shape = (1,), name = "superhost")
input_cat2 = tf.keras.layers.Input(shape = (1,), name = "identity")
input_cat3 = tf.keras.layers.Input(shape = (1,), name = "neighbourhood")
input_cat4 = tf.keras.layers.Input(shape = (1,), name = "room")

x_numeric = tf.keras.layers.Dense(32, name = "encoding_num")(input_num)
x_categoric1 = tf.keras.layers.Embedding(input_dim = 2, output_dim = 3, name = "embedding_cat1")(input_cat1)
x_categoric1 = tf.keras.layers.Reshape(target_shape = (3,), name = "flat_vec1")(x_categoric1)
x_categoric2 = tf.keras.layers.Embedding(input_dim = 2, output_dim = 3, name = "embedding_cat2")(input_cat2)
x_categoric2 = tf.keras.layers.Reshape(target_shape = (3,), name = "flat_vec2")(x_categoric2)
x_categoric3 = tf.keras.layers.Embedding(input_dim = 21, output_dim = 3, name = "embedding_cat3")(input_cat3)
x_categoric3 = tf.keras.layers.Reshape(target_shape = (3,), name = "flat_vec3")(x_categoric3)
x_categoric4 = tf.keras.layers.Embedding(input_dim = 4, output_dim = 3, name = "embedding_cat4")(input_cat4)
x_categoric4 = tf.keras.layers.Reshape(target_shape = (3,), name = "flat_vec4")(x_categoric4)

x_all = tf.keras.layers.Concatenate()([x_numeric, x_categoric1,x_categoric2,x_categoric3,x_categoric4])
x_all = tf.keras.layers.Dense(512, name = "hidden_layer_1", activation="tanh")(x_all)
x_all = tf.keras.layers.Dropout(rate=0.2)(x_all)
x_all = tf.keras.layers.Dense(512, name = "hidden_layer_2", activation="tanh")(x_all)
x_all = tf.keras.layers.Dropout(rate=0.2)(x_all)
x_all = tf.keras.layers.Dense(256, name = "hidden_layer_3", activation="tanh")(x_all)
x_all = tf.keras.layers.Dropout(rate=0.2)(x_all)
x_all = tf.keras.layers.Dense(256, name = "hidden_layer_4", activation="tanh")(x_all)
x_all = tf.keras.layers.Dropout(rate=0.2)(x_all)
x_all = tf.keras.layers.Dense(128, name = "hidden_layer_5", activation="tanh")(x_all)
x_all = tf.keras.layers.Dropout(rate=0.2)(x_all)
x_all = tf.keras.layers.Dense(128, name = "hidden_layer_6", activation="tanh")(x_all)
x_all = tf.keras.layers.Dropout(rate=0.2)(x_all)
x_all = tf.keras.layers.Dense(64, name = "hidden_layer_7", activation="tanh")(x_all)
x_all = tf.keras.layers.Dropout(rate=0.2)(x_all)
x_all = tf.keras.layers.Dense(32, name = "hidden_layer_8", activation="tanh")(x_all)
x_all = tf.keras.layers.Dropout(rate=0.2)(x_all)
x_all = tf.keras.layers.Dense(1, name = "final_layer")(x_all)
model_in = tf.keras.models.Model([input_num, input_cat1, input_cat2, input_cat3, input_cat4], x_all, name = "funct_model_all")
```

```python
print(model_in.summary())
```

```python
model_in.compile("adam", "mean_squared_error", metrics = ["mean_absolute_error"])
```

```python
history = model_in.fit((X[CONTINUAL_COLUMNS], X['host_is_superhost'], X['host_identity_verified'], X['neighbourhood_group_cleansed'], X['room_type']), 
                              y, batch_size = 32, epochs = 150, validation_split=0.2)
```

```python
training_mae = history.history['mean_absolute_error']
test_mae = history.history['val_mean_absolute_error']

epoch_count = range(1, len(training_mae) + 1)

plt.plot(epoch_count, training_mae, 'r--')
plt.plot(epoch_count, test_mae, 'b-')
plt.legend(['Training MAE', 'Test MAE'])
plt.xlabel('Epoch')
plt.ylabel('MAE')
plt.show();
```
